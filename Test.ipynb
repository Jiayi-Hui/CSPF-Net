# CSPF-Net Text Prototype

This notebook shows how to load a Hugging Face text dataset, extract `style + probabilistic + cohesion` features, and train a simple fusion model.

```python
# Run once if your environment is still empty.
# !pip install -r requirements.txt
```

```python
from cspf_text import CSPFTextPipeline, load_text_dataset

bundle = load_text_dataset(
    name="raid",
    split="train",
    sample_size=256,
    text_column="text",
    label_column="label",
)

pipeline = CSPFTextPipeline()
pipeline.fit(bundle.texts, bundle.labels)
pipeline.feature_names_[:10]
```

```python
document = """
Large language models can produce fluent paragraphs very quickly.
However, human writing often contains more uneven rhythm and less uniform phrasing.
""".strip()

result = pipeline.predict_document(document)
result
```
